### Properties of the Transpose of a Matrix

- ($a^t)^t$ = a
- $(a + b)^t$ = $a^t$ + $b^t$
- $(ab)^t$ = $b^ta^t$
- $(ra)^t$ = $ra^t$


##### ($a^t)^t$ = a

```python
a
```

```python
(a.T).T
```

```python
(a.T).T == a
```

#####  $(a + b)^t$ = $a^t$ + $b^t$

```python
(a + b).T
```

```python
a.T + b.T
```

```python
(a + b).T == a.T + b.T
```

##### - $(ab)^t$ = $b^ta^t$

```python
(a.dot(b)).T
```

```python
(b.T).dot(a.T)
```

```python
(a.dot(b)).T == (b.T).dot(a.T)
```
