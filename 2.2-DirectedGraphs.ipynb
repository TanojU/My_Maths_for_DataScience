```python
import networkx as nx
import matplotlib.pyplot as plt
```

### Directed graphs

- Adding and edge from A to C, creates only an edge from A to C
- The edge from C to A has to be explicitly added, it is not implicit

```python
G = nx.DiGraph()

G
```

```python
G.add_node('A')
G.add_node('E')

G.nodes
```

```python
G.add_nodes_from(['B', 'C', 'D'])

G.nodes
```

```python
plt.subplots()

nx.draw(G)
```

```python
def plot_graph(G, node_color, edge_color):
    
    plt.subplots(figsize=(8, 8))
    ax = plt.gca()
    ax.margins(0.20)
    plt.axis("off")

    options = {
        "font_size": 24,
        "with_labels": True,
        "node_size": 1500,
        "node_color": node_color,
        "edgecolors": edge_color,
        "linewidths": 5,
        "width": 5,
    }

    nx.draw_shell(G, **options)
```

```python
plot_graph(G, "cyan", "black")
```

Note that we have an OutEdgeView rather than just a EdgeView

```python
G.add_edge('A', 'B')

G.nodes, G.edges
```

```python
G.add_edge('B', 'A')

G.nodes, G.edges
```

```python
G.add_edge('B', 'C')
G.add_edge('D', 'C')
G.add_edge('C', 'E')
G.add_edge('E', 'B')

G.nodes, G.edges
```

Note the edge between A and B is bi-directional, all other edges are unidirectional

```python
plot_graph(G, "azure", "blue")
```

```python
G.number_of_nodes()
```

Note that the edge from A->B and B->A is counted twice, i.e. there are 2 edges

```python
G.number_of_edges()
```

Create a Graph, and convert to a DiGraph

```python
H = nx.Graph()

H.nodes, H.edges
```

```python
H.add_nodes_from(['X', 'Y', 'Z'])

H.nodes, H.edges
```

Cannot have multiple edges between nodes for Graph as well as DiGraph. NetworkX has a special structure called MultiGraph which supports multiple edges between nodes

```python
H.add_edges_from([('X', 'Z'), ('X', 'Z')])

H.nodes, H.edges
```

```python
H.add_edges_from([('Y', 'Z'), ('X', 'Z'), ('X', 'Y')])

H.nodes, H.edges
```

```python
plot_graph(H, 'magenta', 'purple')
```

Can convert undirected graphs to directed graphs, and vice-versa

```python
H = H.to_directed()

H.nodes, H.edges
```

```python
plot_graph(H, 'yellow', 'red')
```

```python
G.add_nodes_from(H.nodes)

plot_graph(G, 'orange', 'brown')
```

```python
G.add_edges_from(H.edges)

plot_graph(G, 'yellow', 'orange')
```

### Degree of a node

- in-degree is the number of incoming edges
- out-degree is the number of outgoing edges
- degree is the total of incoming and outgoing nodes
- in an undirected graph in-degree == out-degree == degree

***In a DiGraph degree = in-degree + out-degree even though this feels inconsistent***
