```python
import math
import numpy as np

from  mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

from scipy.misc import derivative
```

##### Lets take below equation which has two variables x and y:

f(x,y) = x^2 + y

```python
@np.vectorize
def func(x, y):
    
    return x**2 + y**2 + x*y
```

```python
plt.figure(figsize=(14, 8))

ax = plt.axes(projection='3d')

x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)

X, Y = np.meshgrid(x, y)

Z = func(X, Y)

ax.plot_surface(X, Y, Z, color='purple', antialiased=True, edgecolor='purple')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
```

### Formula for partial derivative

http://www.ltcconline.net/greenl/courses/107/premid3/partial.htm

```python
def get_partial_derivative(func, const_value, x=None, y=None):
    
    h = 0.0001
    
    if x is not None:
        y = const_value
        return (func(x+h, y) - func(x, y)) / h
    
    if y is not None:
        x = const_value
        return (func(x, y+h) - func(x, y)) / h
```

##### Diff at x=1, x=5, x=10

```python
const_value = 1

const_value
```

```python
get_partial_derivative(func, const_value, x=1)
```

```python
get_partial_derivative(func, const_value, x=5)
```

```python
get_partial_derivative(func, const_value, x=10)
```

##### Diff at y=1, y=5, y=10

```python
get_partial_derivative(func, const_value, y=1)
```

```python
get_partial_derivative(func, const_value, y=5)
```

```python
get_partial_derivative(func, const_value, y=10)
```

##### Plotting graphs

```python
plt.figure(figsize=(14, 8))

ax = plt.axes(projection='3d')

x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)

X, Y = np.meshgrid(x, y)

Z0 = func(X, Y)

ax.plot_surface(X, Y, Z0, color='purple', antialiased=True, edgecolor='purple')


Z1 = get_partial_derivative(func, const_value, x=X)
Z2 = get_partial_derivative(func, const_value, y=Y)

ax.plot_surface(X, Y, Z1, color='green', antialiased=True, edgecolor='green')
ax.plot_surface(X, Y, Z2, color='red', antialiased=True, edgecolor='red')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
```

### Using Sympy

```python
import sympy as sp
```

```python
x, y = sp.symbols('x, y')
```

```python
eq = x**2 + y

eq
```

##### Diff with respect to x

```python
diff_wrt_x = sp.diff(eq, x)

diff_wrt_x
```

##### Diff at x=1, x=5, x=10

```python
diff_wrt_x.subs(x, 1)
```

```python
diff_wrt_x.subs(x, 5)
```

```python
diff_wrt_x.subs(x, 10)
```

##### Diff with respect to y

```python
diff_wrt_y = sp.diff(eq, y)

diff_wrt_y
```

##### Diff at y=1, y=5, y=10

```python
diff_wrt_y.subs(y, 1)
```

```python
diff_wrt_y.subs(y, 5)
```

```python
diff_wrt_y.subs(y, 10)
```

## Exponential Function

```python
@np.vectorize
def exp_func(x, y):
    
    return math.exp(x + 0.1*y)
```

##### Diff at x=1, x=5, x=10

```python
get_partial_derivative(exp_func, const_value, x=1)
```

```python
get_partial_derivative(exp_func, const_value, x=5)
```

```python
get_partial_derivative(exp_func, const_value, x=10)
```

##### Diff at y=1, y=5, y=10

```python
get_partial_derivative(exp_func, const_value, y=1)
```

```python
get_partial_derivative(exp_func, const_value, y=5)
```

```python
get_partial_derivative(exp_func, const_value, y=10)
```

```python
plt.figure(figsize=(20,10))

ax = plt.axes(projection='3d')

x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)

X, Y = np.meshgrid(x, y)

Z0 = exp_func(X, Y)

ax.plot_surface(X, Y, Z0, color='purple', antialiased=True, edgecolor='purple')

Z1 = get_partial_derivative(exp_func, const_value, x=X)
Z2 = get_partial_derivative(exp_func, const_value, y=Y)

ax.plot_surface(X, Y, Z1, color='green', antialiased=True, edgecolor='green')
ax.plot_surface(X, Y, Z2, color='red', antialiased=True, edgecolor='red')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
```
